Este projeto tem por objetivo aplicar os cálculos de avaliação a partir de uma matriz de confusão arbitrária.

- (English Translate)

"This project aims to apply evaluation calculations based on an arbitrary confusion matrix."

O primeiro passo é importar as bibliotecas que serão utilizadas ao longo do código


- (English Translate)


The first step is to import the libraries that will be used throughout the code.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score



```

Criando valores arbitrários

(English Translate)

"Creating arbitrary values"


```python
# 1. Defini valores arbitrários
# [[Verdadeiros Negativos, Falsos Positivos],
#  [Falsos Negativos, Verdadeiros Positivos]]
data = np.array([[45, 5],
                 [10, 40]])
```

Criando o gráfico da matriz de confusão

(English Translate)


"Plotting the confusion matrix

```python
# 2. Cria o gráfico
plt.figure(figsize=(6, 4))
sns.heatmap(data, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Negativo', 'Positivo'],
            yticklabels=['Negativo', 'Positivo'])

plt.xlabel('Predito')
plt.ylabel('Real')
plt.title('Matriz de Confusão Arbitrária')
plt.show()
```

Considerando a matriz arbitrária:

[TN, FP]

[FN, TP]

(English Translate)

Considering the arbitrary matrix

```python
# Considerando a matriz arbitrária:
# [[TN, FP],
#  [FN, TP]]
data = np.array([[45, 5],
                [10, 40]])
```

Extração dos valores

(English Translate)


Extracting the values

```python
# Extrair os valores
tp = data[1, 1] # Verdadeiros Positivos
fn = data[1, 0] # Falsos Negativos
```

Calcular a sensibilidade

(English Translate)

Calculating the sensitivity

```python
# Calcular a sensibilidade
sensibilidade = (tp / (tp + fn)) * 100

print(f"Sensibilidade: {sensibilidade:.1f}%")
```

Definindo valores arbitrários

(English Translate)

Defining arbitrary values

```python
# 1. Defina dados arbitrários (0 = Negativo, 1 = Positivo)
y_true = [0, 0, 0, 1, 1, 1, 1, 0] # O que aconteceu de fato
y_pred = [0, 0, 1, 1, 1, 0, 1, 0] # O que o modelo previu


tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
```

Calcular a Especificidade

(English Translate)

Calculating the Specificity

```python
especificidade = (tn / (tn + fp)) * 100
print(f"Especificidade: {especificidade:.1f}%")
```

Calcular a Acurácia

(English Translate)

Calculating the accuracy

```python
# Usando as listas definidas anteriormente
acuracia = (accuracy_score(y_true, y_pred)) * 100
print(f"Acurácia: {acuracia:.1f}%")
```

Calcular a Precisão

(English Translate)

Calculating the Precision

```python
# Usando as listas y_true e y_pred definidas anteriormente
precisao = (precision_score(y_true, y_pred)) * 100
print(f"Precisão: {precisao:.1f}%")
```

Calcular o F-Scre

(English Translate)

Calculating the F-Scre

```python
# Usando as listas y_true e y_pred
f1 = (f1_score(y_true, y_pred)) * 100
print(f"F1-Score: {f1:.1f}%")
```
